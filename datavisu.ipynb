```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import folium
from geopy.extra.rate_limiter import RateLimiter
from geopy.geocoders import Nominatim

# Load and preprocess data
df = pd.read_csv("valeursfoncieres-2023.txt", sep="|")
df = df.drop_duplicates()
df['No voie'] = df['No voie'].fillna(0).astype(int).astype(str)
df['Code postal'] = df['Code postal'].fillna(0).astype(int).astype(str)
df_address = pd.DataFrame({
    'ADDRESS': df['No voie'] + " " + df['Type de voie'] + " " + df['Voie'] + " " + df['Code postal'] + " " + df['Commune'],
    'VF': df['Valeur fonciere'],
    'type': df['Type local']
}).dropna()

# Sample random addresses to reduce geocode calls
df_address = df_address.sample(n=100)

# Set up geocoder with rate limiter
geolocator = Nominatim(user_agent="dvf")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

# Apply geocoding
df_address['location'] = df_address['ADDRESS'].apply(geocode)
df_address['point'] = df_address['location'].apply(lambda loc: tuple(loc.point) if loc else None)
df_address = df_address.dropna(subset=['point'])

# Split point column into latitude, longitude, and altitude columns
df_address[['latitude', 'longitude', 'altitude']] = pd.DataFrame(df_address['point'].tolist(), index=df_address.index)

# Check the DataFrame
print(df_address.head())
```

```python
# for numerical computing
import numpy as np
# for dataframes
import pandas as pd
# for easier visualization
%matplotlib inline
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
```

```python
folium_map = folium.Map(location=[46.2276, 2.2137],
                        zoom_start=6,
                        tiles='CartoDB dark_matter')


FastMarkerCluster(data=list(zip(df_address['latitude'].values, df_address['longitude'].values,df_address['VF'].values))).add_to(folium_map)
folium.LayerControl().add_to(folium_map)
folium_map
```

```python
m = folium.Map(location=[46.2276, 2.2137],
                        zoom_start=6,
                        tiles='CartoDB dark_matter')

# I can add marker one by one on the map
for i in range(0,len(df_address)):
    popup_text = folium.Html('<b>'+'Address: '+'</b>' + df_address.iloc[i]['ADDRESS'] + '<br>'
                   '<b>'+'Valeur Fonciere: '+'</b>' +df_address.iloc[i]['VF'] + '<br>'
                   '<b>'+'Property Type: ' +'</b>'+str(df_address.iloc[i]['type']), script=True)
    popup = folium.Popup(popup_text, max_width=400,min_width=400)
    tooltip='click me!'
    folium.Marker([df_address.iloc[i]['latitude'], df_address.iloc[i]['longitude']], popup=popup,tooltip=tooltip).add_to(m)
folium.LayerControl().add_to(m)
```
